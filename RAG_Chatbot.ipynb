```python
import bs4
from langchain import hub
from langchain_chroma import Chroma
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

# Document loader

```python
from langchain_unstructured import UnstructuredLoader

file_paths = [
    "/content/Arena Breakout.docx"

]

loader = UnstructuredLoader(file_paths)
docs = loader.load()
```

```python
from langchain_nvidia_ai_endpoints import ChatNVIDIA

client = ChatNVIDIA(
  model="meta/llama-3.1-405b-instruct",
  api_key=NVIDIA_API_KEY,
  temperature=0.2,
  top_p=0.7,
  max_tokens=1024,
)
```

# Text splitters

```python
#from langchain_nvidia_ai_endpoints import NVIDIAEmbeddings
```

```python
#embedder = NVIDIAEmbeddings(model="NV-Embed-QA", api_key=NVIDIA_API_KEY)
```

```python
from langchain_huggingface.embeddings import HuggingFaceEmbeddings

embedder = HuggingFaceEmbeddings(model_name="BAAI/bge-small-en-v1.5")
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)

splits = text_splitter.split_documents(docs)
```

```python
splits[0]
```

```python
for idx, text in enumerate(splits):
    text.metadata["id"] = idx
```

# Vector store and Reranking

```python
from langchain_community.vectorstores.utils import filter_complex_metadata

vectorstore = Chroma.from_documents(documents=filter_complex_metadata(splits), embedding=embedder)
```

```python
retriever = vectorstore.as_retriever(search_kwargs={"k": 20})
```

```python
# Helper function for printing docs
def pretty_print_docs(docs):
    print(
        f"\n{'-' * 100}\n".join(
            [
                f"Document {i+1}:\n\n{d.page_content}\nMetadata: {d.metadata}"
                for i, d in enumerate(docs)
            ]
        )
    )
```

```python
query = "cách để chơi arena breakout"
docs = retriever.invoke(query)
pretty_print_docs(docs)
```

```python
from langchain.retrievers.contextual_compression import ContextualCompressionRetriever
from langchain_cohere import CohereRerank
from langchain_community.llms import Cohere

compressor = CohereRerank(cohere_api_key=COHERE_API_KEY, model="rerank-english-v3.0")
compression_retriever = ContextualCompressionRetriever(
    base_compressor=compressor, base_retriever=retriever
)

compressed_docs = compression_retriever.invoke(
    "Cách chơi arena break out"
)
pretty_print_docs(compressed_docs)
#print([doc.metadata["id"] for doc in compressed_docs])
```

```python
prompt = hub.pull("rlm/rag-prompt")
```

```python
prompt
```

```python
prompt.messages[0].prompt.template = "You are an assistant for question-answering tasks of 'Arena Breakout' game. Use the following pieces of retrieved context to answer the question. If you don't know the answer, just say that you don't know. Use three sentences maximum and keep the answer concise.\nQuestion: {question} \nContext: {context} \nAnswer:"
```

```python
prompt
```

```python
def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)


rag_chain = (
    {"context": compression_retriever | format_docs, "question": RunnablePassthrough()}
    | prompt
    | client
    | StrOutputParser()
)
```

```python
for chunk in rag_chain.stream("cách chơi game"):
  print(chunk, end="")
```

```python
import gradio as gr

def rag_chatbot(message, history):
    response = ""
    for chunk in rag_chain.stream(message):
      response += chunk
      yield response

gr.ChatInterface(rag_chatbot, type="messages").launch()
```
